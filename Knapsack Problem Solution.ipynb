```python
inputString = input("Enter input file:")
outputString = input("Enter output file:")
inputFile = open(inputString, "r")
lines = inputFile.readlines()
```

```python
status = 0
m = 0
n = 0
n_values = []
m_values = []
mxn = [[]]
i = 0
j = 0
for line in lines:
    textArray = line.strip().split()
    if(status == 0):
        m = int(textArray[0])
        n = int(textArray[1])
        status += 1
    elif(status == 1):
        for text in textArray:
            n_values.append(int(text))
        if(len(n_values) >= n):
            status += 1
    elif(status == 2):
        for text in textArray:
            m_values.append(int(text))
        if(len(m_values) >= m):
            status += 1
    elif(status == 3):
        for text in textArray:
            mxn[i].append(int(text))
            j += 1
        if(j >= n):
            mxn.append([])
            i += 1
            j = 0
        if(i >= m):
            status += 1
```

```python
outputVariables = [0 for x in range(0, n)]
```

```python
#Creates an array for value efficiency function from the input data
valueArray = [[]]
def createValueArray():
    for x in range(0, n):
        value = 0
        for y in range(0, m):
            value = value + mxn[y][x]
        valueArray[x].append(x)
        valueArray[x].append(value)
        if(x < n-1): valueArray.append([])
    return valueArray
```

```python
def takeSecond(element):
    return element[1]
```

```python
#value efficiency function takes two arguments array and length and then calculates the sum of every knapsack weight for each item.
#After calculations the array will be send to create a point array
sortedValueArray = [[]]
def calculateValueEfficiency(array, length):
    global sortedValueArray
    sortedValueArray.clear()
    sortedValueArray = [[]]
    i = 0
    value = 0
    for i in range(0, length):
        if(array[i][1] != 0) : value = n_values[array[i][0]] / array[i][1]
        else: value = n_values[i]
        sortedValueArray[i].append(array[i][0])
        sortedValueArray[i].append(value)
        if(i < length - 1): sortedValueArray.append([])
    sortedValueArray.sort(key=takeSecond, reverse=True)
    return sortedValueArray
```

```python
#Finds the values that have a weight more then empty knapsack capacity
weightArray = [[]]
def calculateLowerWeight(value, weightIndexes):
    global weightArray
    weightArray.clear()
    weightArray = [[]]
    i = 0
    newArray = True
    for weightIndex in weightIndexes:
        isHigher = False
        for index in range(0, m):
            if(mxn[index][weightIndex] >= value):
                isHigher = True
        if(isHigher == False):
            if(newArray == False): weightArray.append([])
            newArray = False
            weightArray[i].append(weightIndex)
            weightArray[i].append(mxn[0][weightIndex] + mxn[1][weightIndex])
            i += 1
    return weightArray
```

```python
#Not an efficienct approach, not used in main program
sortedDistanceArray = [[]]
def calculateDistance():
    i = 0
    for i in range(0, n):
        sortedDistanceArray[i].append(i)
        sortedDistanceArray[i].append(abs(mxn[0][i] - mxn[1][i]))
        if(i < n-1): sortedDistanceArray.append([])
    sortedDistanceArray.sort(key=takeSecond)
    return sortedDistanceArray
```

```python
#Not an efficienct approach, not used in main program
def calculateWeight():
    averageWeight = 0
    for i in range(0, n):
        averageWeight = averageWeight + mxn[0][i] + mxn[1][i]
    averageWeight = averageWeight / n
    i = 0
    for i in range(0, n):
        reducedPoint = 0
        sortedWeightArray[i].append(i)
        if(mxn[0][i] > averageWeight - averageWeight / 2): reducedPoint -= 10
        if(mxn[1][i] > averageWeight - averageWeight / 2): reducedPoint -= 10
        sortedWeightArray[i].append(reducedPoint)
        if(i < n-1): sortedWeightArray.append([])
    return sortedWeightArray
```

```python
#Attach efficiency score to the index point array
indexPointArray = [[]]
def givePoints():
    global indexPointArray
    indexPointArray.clear()
    indexPointArray = [[]]
    i = 0
    count = 0
    for i in range(0, n):
        for valueItem in sortedValueArray:
            if(valueItem[0] == i):
                indexPointArray[count].append(valueItem[0])
                indexPointArray[count].append(valueItem[1])
                count += 1
                if(count < len(sortedValueArray)): indexPointArray.append([])
                break
    indexPointArray.sort(key=takeSecond, reverse=True)
    return indexPointArray
```

```python
createValueArray()
calculateValueEfficiency(valueArray, n)
givePoints()
```

```python
#Calculate for the first time to find the total value
totalValue = 0
index = 0
m_item_weights = [0 for x in range(0,m)]
weightIndexes = [x for x in range(0,n)]
endLoop = False
for itemPoint in indexPointArray:
    for i in range(0,m):
        if(m_item_weights[i] + mxn[i][itemPoint[0]] > m_values[i]):
            index = i
            endLoop = True
            break
        else:
            m_item_weights[i] = m_item_weights[i] + mxn[i][itemPoint[0]]
    if(endLoop): break
    else:
        totalValue = totalValue + n_values[itemPoint[0]]
        outputVariables[itemPoint[0]] = 1
        weightIndexes.remove(itemPoint[0])
print(totalValue)
```

```python
#Calculate until the knapsack can be filled and find the maximum value and print the values to the output file
while(1):
    remainingWeight = 0
    for i in range(0,m):
        if(remainingWeight == 0): remainingWeight = m_values[i] - m_item_weights[i]
        if(m_values[i] - m_item_weights[i] < remainingWeight): remainingWeight = m_values[i] - m_item_weights[i]
    calculateLowerWeight(remainingWeight, weightIndexes)
    if(len(weightArray) < 2 ): break
    calculateValueEfficiency(weightArray, len(weightArray))
    givePoints()
    endLoop = False
    index = 0
    for itemPoint in indexPointArray:
        for i in range(0,m):
            if(m_item_weights[i] + mxn[i][itemPoint[0]] >= m_values[i]):
                index = i
                endLoop = True
                break
            else:
                m_item_weights[i] = m_item_weights[i] + mxn[i][itemPoint[0]]
        if(endLoop): break
        else:
            totalValue = totalValue + n_values[itemPoint[0]]
            outputVariables[itemPoint[0]] = 1
            weightIndexes.remove(itemPoint[0])
print(totalValue)
print(outputVariables)
newString = str(totalValue) + "\n"
file2 = open(outputString, "a")
file2.write(newString)
for i in range(0,n):
    newString = str(outputVariables[i]) + "\n"
    file2.write(newString)
```
